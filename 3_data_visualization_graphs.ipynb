```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the 'tips' dataset from seaborn
tips = sns.load_dataset('tips')

# Set up the plot style
sns.set(style="whitegrid")
```

```python
# 1. Histogram of total bill
plt.figure(figsize=(3, 2))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Total Bill', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# 2. Boxplot for total bill by day of the week
plt.figure(figsize=(3, 2))
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm')
plt.title('Boxplot of Total Bill by Day', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Total Bill ($)', fontsize=12)
plt.show()
```

```python
# 3. Scatter plot of total bill vs tip
plt.figure(figsize=(7, 5))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='time', palette='muted', style='sex', s=100)
plt.title('Scatterplot of Total Bill vs Tip', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.show()
```

```python
# 4. Pairplot for exploring relationships between all numerical variables
sns.pairplot(tips, hue='sex', palette='Set1')
plt.suptitle('Pairplot of Numerical Variables by Gender', y=1.02, fontsize=16)
plt.show()
```

```python
# 5. Countplot of gender distribution
plt.figure(figsize=(3, 2))
sns.countplot(x='sex', data=tips, palette='pastel')
plt.title('Countplot of Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# 6. Violin plot of tips by time of day
plt.figure(figsize=(3, 2))
sns.violinplot(x='time', y='tip', data=tips, palette='muted')
plt.title('Violin Plot of Tips by Time of Day', fontsize=16)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)
plt.show()
```

```python
# 7. Bar plot of average tip by day of the week
plt.figure(figsize=(3, 2))
sns.barplot(x='day', y='tip', data=tips, palette='Blues_d')
plt.title('Average Tip by Day of the Week', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Average Tip ($)', fontsize=12)
plt.show()
```

```python
# 8. FacetGrid for day and time breakdown
g = sns.FacetGrid(tips, col="day", row="time", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip", alpha=.7, color='blue')
g.set_axis_labels("Total Bill ($)", "Tip ($)")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('FacetGrid of Tips by Day and Time', fontsize=16)
plt.show()
```
